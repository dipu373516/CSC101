```python
contacts = {
    "Alice": "0400 123 456",
    "Bob": "0411 000 999",
    "Charlie": "0422 111 222"
}
```

```python
contacts["Dipu"] = "0433 222 333"
```

```python
contacts["Bob"] = "0499 888 777"
```

```python
for name, phone in contacts.items():
    print(f"{name}: {phone}")
```

```python
contacts = {
    "Alice": {"phone": "0400 123 456", "email": "alice@example.com"},
    "Bob": {"phone": "0499 888 777", "email": "bob@example.com"},
    "Dipu": {"phone": "0433 222 333", "email": "rathul@example.com"}
}
```

```python
for name, details in contacts.items():
    print(f"Name: {name}")
    print(f"Phone: {details['phone']}")
    print(f"Email: {details['email']}")
    print()
```

```python
search_name = input("Enter a name: ")

if search_name in contacts:
    print("Phone:", contacts[search_name]["phone"])
    print("Email:", contacts[search_name]["email"])
else:
    print("Contact not found")
```
